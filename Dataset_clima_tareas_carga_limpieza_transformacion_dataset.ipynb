## Se carga el archivo "datos_clima.csv" en un dataframe y se muestra su contenido

### Se lee el archivo "datos_clima.csv" y se guarda en la variable clima

```python
clima = pd.read_csv("datos_clima.csv")
clima
```

## Se muestra las columnas con sus tipos de datos del dataframe

```python
clima.info()
```

## Eliminación de las columnas que tienen todos los valores nulos

### Se eliminan columnas que tengan todos los valores nulos y se guarda el dataframe obtenido en una nueva variable

```python
climaSinValsNull = clima.dropna(axis="columns", how="all")
```

```python
climaSinValsNull
```

## Identificación de filas duplicadas y eliminación de las mismas

### Identificación de filas duplicadas

```python
climaSinValsNull.duplicated()
```

### Eliminación de filas duplicadas y asignación en un nuevo dataframe

```python
climaSinDuplic = climaSinValsNull.drop_duplicates()
```

```python
climaSinDuplic
```

## Identificación de valores nulos

```python
climaSinDuplic.isna()
```

## Reemplazo valores nulos por cero

### Reemplazo de valores nulos por ceros y asignación en un nuevo dataframe

```python
climaReempCeros = climaSinDuplic.fillna(0)
```

### Verificación de reemplazo de valores nulos por ceros

```python
climaReempCeros.isna()
```
