## Se descarga archivo subido previamente a drive con permisos necesarios para descargarlo y se carga el dataset a un dataframe

```python
titanic = pd.read_csv(filename)
```

## Se indica cantidad de filas y columnas del dataframe

```python
cant_filas = titanic.index.size
cant_cols = titanic.columns.size

print("La cantidad de filas es: ", cant_filas, "\nLa cantidad de columnas es: ", cant_cols)
```

## Verificación de valores nulos en el dataFrame y se muestra la cantidad de valores nulos por columna

```python
titanic.isna()
```

```python
titanic.isna().sum()
```

## Se completa los valores nulos en la columna 'Age' con la media

```python
age_media = titanic["Age"].mean()
print(age_media)
```

```python
titanic["Age"].fillna(age_media, inplace=True)
```

```python
titanic.isna().sum()
```

## Se muestra un gráfico de histograma que muestra la distribución de edades de los pasajeros en el Titanic

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(x=titanic["Age"], bins= 8)
plt.style.use('ggplot')
plt.xlabel('Edades')
plt.ylabel('Frecuencia de edades')
plt.title('Distribución de edades de los pasajeros del Titanic')
plt.show()
```

## Se muestra la tasa de supervivencia de los pasajeros

```python
survived_mean = titanic["Survived"].mean()
print("La tasa promedio de supervivencia fue: ", round(survived_mean*100, ndigits=2), " %")
```
